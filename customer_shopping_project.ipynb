```python
category_median = df.groupby('Category')['Review Rating'].transform('median')
```

```python
df['Review Rating'] = df['Review Rating'].fillna(category_median)
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" ","_")
```

```python
df.columns
```

```python
df.head()
```

```python
df= df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# creating a colum of age_group
labels = ["Young Adult", "Adult", "Middle-aged", "Senior"]
df["age_group"] = pd.qcut(df["age"], q=4, labels=labels)
```

```python
df[["age","age_group"]].head(10)
```

```python
# creating a colum purchases_frequency_days
frequency_mapping = {
    "Fortnightly": 14,
    "Weekly": 7,
    "Monthly": 30,
    "Quarterly": 90,
    "Annually": 365,
    "Bi-Weekly" : 14,
    "Every 3 Months": 90
}

df["purchases_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchases_frequency_days","frequency_of_purchases"]].head(5)
```

```python
print(df["purchases_frequency_days"].isnull().sum())
```
